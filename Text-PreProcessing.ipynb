```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

# Remove Puncuations,numbers,converting to lowercase,Stemming

```python
review=data['body'][0]
```

```python
review
```

```python
data1=[]
for i in range(0,1000):
    review=data["body"][i]
    review=re.sub('[^a-zA-Z]',' ',review)
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    data1.append(review)
```

```python
data1
```
